```python
import torch
import torch.nn.functional as F

# Toy sentence embeddings (5 tokens, embedding dim = 4)
torch.manual_seed(42)
embeddings = torch.randn(5, 4)

# Create Query, Key, Value matrices
W_q = torch.randn(4, 4)
W_k = torch.randn(4, 4)
W_v = torch.randn(4, 4)

Q = embeddings @ W_q
K = embeddings @ W_k
V = embeddings @ W_v

# Scaled Dot-Product Attention
scores = Q @ K.T / (K.size(-1) ** 0.5)
attention_weights = F.softmax(scores, dim=-1)

output = attention_weights @ V

print("Attention Scores (before softmax):")
print(scores)

print("\nAttention Weights (after softmax):")
print(attention_weights)

print("\nFinal Attention Output:")
print(output)
```
