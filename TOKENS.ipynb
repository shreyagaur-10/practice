```python
import tiktoken

# Choose model encoding
encoding = tiktoken.encoding_for_model("gpt-4o-mini")

def count_tokens(text):
    tokens = encoding.encode(text)
    return len(tokens)

sample_text = """
Large Language Models are powerful tools,
but token usage directly affects cost.
"""

print("Token count:", count_tokens(sample_text))
```

```python
texts = [
    "Explain LLMs",
    "Explain Large Language Models in detail with examples",
    "Explain Large Language Models in detail with examples and architecture diagrams"
]

for t in texts:
    print(count_tokens(t), "→", t)
```
